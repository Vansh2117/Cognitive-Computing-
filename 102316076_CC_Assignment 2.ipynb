```python
# Q.1
L = [10, 20, 30, 40, 50, 60, 70, 80]

L.append(200)
L.append(300)

L.remove(10)
L.remove(30)

L.sort()
print("Ascending Order:", L)

L.sort(reverse=True)
print("Descending Order:", L)
```

```python
# Q.2
scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45)

highest_score = max(scores)
highest_index = scores.index(highest_score)
print(f"Highest Score: {highest_score}, Index: {highest_index}")

lowest_score = min(scores)
count_lowest = scores.count(lowest_score)
print(f"Lowest Score: {lowest_score}, Count: {count_lowest}")

reversed_list = list(scores[::-1])
print("Reversed List:", reversed_list)

search_score = int(input("Enter the score to search: "))
if search_score in scores:
    first_occurrence = scores.index(search_score)
    print(f"Score {search_score} found at index: {first_occurrence}")
else:
    print(f"Score {search_score} is not present in the tuple.")
```

```python
# Q.3
import random
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

random_numbers = [random.randint(100, 900) for _ in range(100)]

odd_numbers = [num for num in random_numbers if num % 2 != 0]
even_numbers = [num for num in random_numbers if num % 2 == 0]
prime_numbers = [num for num in random_numbers if is_prime(num)]

print("Random Numbers:", random_numbers)
print(f"Odd Numbers ({len(odd_numbers)}): {odd_numbers}")
print(f"Even Numbers ({len(even_numbers)}): {even_numbers}")
print(f"Prime Numbers ({len(prime_numbers)}): {prime_numbers}")
```

```python
# Q.4
A = {34, 56, 78, 90}
B = {78, 45, 90, 23}

unique_scores = A.union(B)
print("Unique scores (Union):", unique_scores)

common_scores = A.intersection(B)
print("Common scores (Intersection):", common_scores)

exclusive_scores = A.symmetric_difference(B)
print("Exclusive scores (Symmetric Difference):", exclusive_scores)

is_A_subset_B = A.issubset(B)
is_B_superset_A = B.issuperset(A)
print(f"Is A a subset of B? {is_A_subset_B}")
print(f"Is B a superset of A? {is_B_superset_A}")

X = int(input("Enter the score to remove from A: "))
if X in A:
    A.remove(X)
    print(f"Score {X} removed from A. Updated A:", A)
else:
    print(f"Score {X} is not present in set A.")
```
