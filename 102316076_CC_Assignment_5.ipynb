```python
import numpy as np
import matplotlib.pyplot as plt
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
sum_all = np.sum(gfg)
print("Sum of all elements:", sum_all)
sum_rows = np.sum(gfg, axis=1)
print("Row-wise sum:", sum_rows)
sum_cols = np.sum(gfg, axis=0)
print("Column-wise sum:", sum_cols)
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
print("Sorted array:", sorted_array)
sorted_indices = np.argsort(array)
print("Indices of sorted array:", sorted_indices)
smallest_4 = np.sort(array)[:4]
print("4 smallest elements:", smallest_4)
largest_5 = np.sort(array)[-5:]
print("5 largest elements:", largest_5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Q2
array_b = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array_b[array_b == array_b.astype(int)]
print("Integer elements:", integer_elements)
float_elements = array_b[array_b != array_b.astype(int)]
print("Float elements:", float_elements)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Q3
first_initial, last_initial = ord('A'), ord('B')
X = first_initial + last_initial
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales array:", sales)
tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 - tax_rate)
print("Sales after tax:", sales_after_tax)
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Discounted sales:", discounted_sales)
sales_matrix = np.tile(sales, (3, 1))
week_adjustment = np.array([1.0, 1.02, 1.04])[:, np.newaxis]
weekly_sales = sales_matrix * week_adjustment
print("Weekly adjusted sales:\n", weekly_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Q4
x = np.linspace(-10, 10, 100)

y_square = x**2
y_sin = np.sin(x)
y_exp = np.exp(x)
y_log = np.log(np.abs(x) + 1)
plt.figure(figsize=(8, 6))
plt.plot(x, y_square, label='y = x^2')
plt.plot(x, y_sin, label='y = sin(x)')
plt.plot(x, y_exp, label='y = e^x')
plt.plot(x, y_log, label='y = log(|x| + 1)')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Mathematical Functions Plot")
plt.legend()
plt.grid()
plt.show()
```
